```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
flight_df=sns.load_dataset('flights')
flight_df
```

```python
flight_df = flight_df.pivot(index="month", columns="year", values="passengers")
flight_df # need to use 3 parameters index,column, values in pivot .
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(flight_df) #in heatmap- lighter the colour- larger the value
ax                         #darker the colour, smaller the value
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(flight_df,annot=True ,fmt='d') #it will show the annotation above the box
ax
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(flight_df,annot=True ,fmt='d',linecolor='k', linewidths='5') #k=black color, b=blue, linewidth= foe line width
ax
```

```python
plt.figure(figsize=(12,6))
ax= sns.heatmap(flight_df,annot=True ,fmt='d',linecolor='k', linewidths='5',cmap='Blues') #cmap- colour map - all plotes becoming a tins of that colour
ax
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flight_df,cbar=False)
```

```python
grid_kws = {"height_ratios": (.4, .05), "hspace": .4}  # hspace is correct- horizontal space, height ratios- if we reduce it height/distance between 2 points will reduce
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)  # here it is plotting 2 subplots, gridspace_kws is a keyword
ax = sns.heatmap(flight_df, cbar_kws={"orientation": "horizontal"}, ax=ax, cbar_ax=cbar_ax)
```

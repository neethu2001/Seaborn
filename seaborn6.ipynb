```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#pair plot-
penguins=sns.load_dataset('penguins')
penguins.head()
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins)
sns.set_style('darkgrid')
plt.style.use('dark_background') #giving dark background
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,hue='sex',palette='Reds') #hue differentiste the data
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,hue='sex',palette='Blues')
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,kind='kde',palette='greens') # kde= kernal density
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,kind='hist',palette='Greens')# hhere it is drawing histogram
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,hue='species',diag_kind='hist',palette='rainbow') # diagonal kind of histogram- diagonals plotted with histogram
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,corner=True)# orner is true means-plotting only the lower part
```

```python
plt.figure(figsize=(12,12))
sns.pairplot(penguins,hue='species',markers=["o","s","D"],palette='inferno')
#here it is representing as marker type, are marker styles used in Matplotlib to represent the data points in the plots
#"o" - Represents a circle, "s"- square, "D"- diamond
```

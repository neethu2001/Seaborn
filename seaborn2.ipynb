```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/hr_data.csv')
df
```

```python
sns.distplot(df['time_spend_company'])
```

```python
sns.distplot(df['left'])
```

```python
sns.distplot(df['average_montly_hours'])
```

```python
#histogram rep data distribution by
#forming bins along with the range of data
#drawing bars shows the no of observations
#that fall in each bin
```

```python
df.describe() #min=2, max=10
```

```python
bins=[2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'],bins=bins)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,
             kde=False)
#kernal density estimate
```

```python
sns.distplot(df['time_spend_company'],bins=bins,
             hist=False)
```

```python
sns.distplot(df['time_spend_company'],bins=bins,
             rug=True) #here datapoints drawing more clearly
```

```python
sns.distplot(df['time_spend_company'],bins=bins,
             rug=True)
plt.xticks(bins)#here we will get that nos in x axis more precisely
```

```python
sns.distplot(df['time_spend_company'], bins=bins,
             rug=True,
             # Disable kernel density estimation
             hist_kws={'color': 'red', 'edgecolor': 'blue', 'linewidth': 3, 'alpha': 0.5})
             # Customizing histogram appearance
              #key word arguments-kws, alpha is the transparency

plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins,
             rug=True,

             hist_kws={'color': 'red', 'edgecolor': 'blue', 'linewidth': 3, 'alpha': 0.5},
             # Customizing histogram appearance
              #key word arguments-kws, alpha is the transparency
             kde_kws={'color': 'orange', 'linewidth': 2})

plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins,
             rug=True,
             color='green')
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins,
             rug=True,
             color='green',
             vertical=True) #Change it into vertical
plt.xticks(bins)
```
